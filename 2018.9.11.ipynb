## EP:
- 通过math库，写一个程序，使得用户输入三个顶点(x,y)返回三个角度
- 注意：Python计算角度为弧度制，需要将其转换为角度
<img src="../Python/Photo/18.png">

```python
import math
x1,y1 = eval(input('输入（x1,y1）:'))
x2,y2 = eval(input('输入（x2,y2）:'))
x3,y3 = eval(input('输入（x3,y3）:'))
a = math.sqrt( math.pow((x3 - x2),2) + math.pow((y2 - y3),2) )
b = math.sqrt( math.pow((x3 - x1),2) + math.pow((y3 - y1),2) )
c = math.sqrt( math.pow((x2 - x1),2) + math.pow((y2 - y1),2) )
A = math.acos((a * a - b * b - c * c) / (-2 * b * c))
B = math.acos((b * b - a * a - c * c) / (-2 * a * c))
C = math.acos((c * c - b * b - a * a) / (-2 * a * b))
print(math.degrees(A))
print(math.degrees(B))
print(math.degrees(C))
```

## 实例研究：最小数量硬币
- 开发一个程序，让用户输入总金额，这是一个用美元和美分表示的浮点值，返回一个由美元、两角五分的硬币、一角的硬币、五分硬币、以及美分个数
<img src="../Python/Photo/22.png"></img>

```python
num = eval(input('输入一个十进制带小数点的数字：'))
fenShu = num * 100
meiYuan = int(fenShu / 100)
shengYuFenShu = fenShu % 100
liangJiaoWuFen = int(shengYuFenShu / 25)
shengYuFenShu = shengYuFenShu % 25
yiJiao = int(shengYuFenShu / 10)
shengYuFenShu = shengYuFenShu % 10
wuFen = int(shengYuFenShu / 5)
shengYuFenShu = int(shengYuFenShu % 5)
print('美元：' + str(meiYuan))
print('两角五分：' + str(liangJiaoWuFen))
print('一角：' + str(yiJiao))
print('五分：' + str(wuFen))
print('美分：' + str(shengYuFenShu))
```
