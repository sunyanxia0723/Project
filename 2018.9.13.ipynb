# 验证码
- 随机产生四个字母的验证码，如果正确，输出验证码正确。如果错误，产生新的验证码，用户重新输入。
- 验证码只能输入三次，如果三次都错，返回“别爬了，我们小网站没什么好爬的”
- 密码登录，如果三次错误，账号被锁定


```python
# 验证码：纯数字
import random
i = 0
while i < 3:
    yzm = str(random.randint(1000,9999))
    print('验证码是:',yzm)
    input_yzm = eval(input('请输入验证码：'))
    if yzm == input_yzm:
        print('验证码正确')
        break
    else:
        print('提示：输入验证码错误，请再次输入')
        i += 1
#     if i ==3:
#         print('验证码次数超限')
else:
    print('验证码次数超限')
```

```python
# 验证码：纯字母
import random
i = 0
while i < 3 :
    yzm1 = chr(random.randint(97,122))
    yzm2 = chr(random.randint(97,122))
    yzm3 = chr(random.randint(97,122))
    yzm4 = chr(random.randint(97,122))
    yzm = yzm1 + yzm2+ yzm3 + yzm4
    print('验证码是:',yzm)
    input_yzm = input('请输入验证码：')
    if yzm == input_yzm:
        print('验证码正确')
        break
    else:
        print('提示：输入验证码错误，请再次输入')
        i +=1
#     if i ==3:
#         print('验证码次数超限')
else:
    print('验证码次数超限')
```

```python
# 验证码：字母+数字
import random
i = 0
while i < 3:
    list_yzm = ['1','2','3','4','5','6','7','8','9','0','q','w','e','r','t','y','u','i','o','p','a','s','d','f','g','h','j','k','l','z','x','c','v','n','m']
    random.shuffle(list_yzm)
    yzm = ''.join(list_yzm[0:4])
    print('验证码是:',yzm)
    input_yzm = input('请输入验证码：')
    
    if yzm == input_yzm:
        print('验证码正确')
        break
    else:
        print('提示：输入验证码错误，请再次输入')
        i +=1
#     if i ==3:
#         print('验证码次数超限')
else:
    print('验证码次数超限')
```

## 嵌套循环
- 一个循环可以嵌套另一个循环
- 每次循环外层时，内层循环都会被刷新重新完成循环
- 也就是说，大循环执行一次，小循环会全部执行一次
- 注意：
> - 多层循环非常耗时
  - 最多使用3层循环

```python
for i in range(10):
    for j in range(10):
        for k in range(10):
            print(i,j,k)
```
